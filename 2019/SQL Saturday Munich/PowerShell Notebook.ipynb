# How many dbatools commands ?

Let's see how many commands there are in dbatools

```powershell
(Get-Command -Module dbatools -CommandType Function ).Count
```

# Lets list the commands

```powershell
## Lets look at the commands
Get-Command -Module dbatools 
```

# Finding dbatools commands

There are so many commands in dbatools that we had to wrote a command to find commands!!

We called it `Find-DbaCommand`

```powershell
Find-DbaCommand -Tag Backup
```

```powershell
Find-DbaCommand -Tag Restore
```

```powershell
Find-DbaCommand -Tag Migration
```

```powershell
Find-DbaCommand -Tag Agent
```

```powershell
Find-DbaCommand -Pattern User 
```

```powershell
Find-DbaCommand -Pattern linked
```

# How do we use commands?

**ALWAYS ALWAYS** use `Get-Help`

```powershell
Get-Help Test-DbaLinkedServerConnection -Full
```

# Here a neat trick

This is how you can use `Out-GridView` to select a command and get the help

```powershell
Find-DbaCommand -Pattern linked | Out-GridView -PassThru | Get-Help -Full 
```
